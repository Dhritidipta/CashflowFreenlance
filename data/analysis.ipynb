```python
import requests
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
from pyspark.sql import SparkSession
from pyspark.sql.functions import when
from pyspark.sql import functions as F
from pyspark.sql.types import FloatType
 
# Initialize a Spark session
spark = SparkSession.builder \
    .appName("Create DataFrame Example") \
    .getOrCreate()

companies_all_data = get_all_companies_df()
exchange_all_data = get_exchange_rates_df()
sepa_all_data = get_all_sepa_transactions_df()
swift_all_data = get_all_swift_transactions_df()
 
# Create a DataFrame
all_companies_df_spark = spark.createDataFrame(companies_all_data)
exchange_df_spark = spark.createDataFrame(exchange_all_data)
sepa_df_spark = spark.createDataFrame(sepa_all_data)
swift_df_spark = spark.createDataFrame(swift_all_data)

# Show the DataFrame
# all_companies_df_spark.show()
all_companies_df_spark.describe()
```

```python
from pyspark.sql.functions import col
 
# Assuming `swift_df_spark` and `exchange_df_spark` are already defined
 
# Join the SWIFT transactions with currency rates

joined_df = swift_df_spark.join(exchange_df_spark, on="currency", how="inner")
 
# Convert amount to EUR

converted_df = joined_df.withColumn("amount", col("amount").cast("float") / col("eur_rate"))
 
# Select relevant columns and rename beneficiary to receival

swift_final_df = converted_df.select("id", "sender", 

                                 col("beneficiary").alias("receiver"),  # Rename here

                                 "amount", "currency", "timestamp")
 
# Show the final DataFrame with the renamed column

swift_final_df.show()

swift_final_df.count()
```

```python
from pyspark.sql.functions import regexp_replace, col
 
# Assuming df_spark is your DataFrame and "iban_column" is the column to clean
 
# Use regexp_replace to remove square brackets and single quotes
cleaned_all_companies_df = all_companies_df_spark.withColumn("ibans", regexp_replace(col("ibans"), "[\\[\\]' ]", ""))
 
# Show the result
cleaned_all_companies_df.show()
cleaned_all_companies_df.describe()
```

```python
final_merged_df.join(cleaned_all_companies_df, on="ibans", how="outer")
```

```python
##sample account number which will be dynamic
account_number = "NL28SZUK8962850954"


# Filter the DataFrame where the account number matches either 'Sender' or 'Receiver'
filtered_df = final_merged_df.filter((col("Sender") == account_number) | (col("receiver") == account_number))

# Show the filtered DataFrame
filtered_df.show(truncate=False)

```

```python

# Convert 'amount' column to float (if it's not already in numerical format)
df = filtered_df.withColumn("amount", col("amount").cast(FloatType()))

# Create a new column 'balance_change' to reflect whether to add or subtract based on 'Sender' or 'receiver'
df_with_balance = df.withColumn(
    "balance_change",
    when(col("Sender") == account_number, -col("amount"))  # Subtract if account is the sender
    .otherwise(col("amount"))  # Add if account is the receiver
)

# Sum up the 'balance_change' to get the account balance
account_balance = df_with_balance.agg({"balance_change": "sum"}).collect()[0][0]

# Print the final account balance
print(f"Account balance for {account_number}: {account_balance}")
```

```python

# Join the DataFrames on ibans with sender and receiver
# We will perform two separate joins: one for sender and one for receiver
sender_join = final_merged_df.join(cleaned_all_companies_df, final_merged_df.Sender == cleaned_all_companies_df.ibans, "inner") \
    .select(cleaned_all_companies_df.name, final_merged_df.amount, final_merged_df.currency, final_merged_df.timestamp)

receiver_join = final_merged_df.join(cleaned_all_companies_df, final_merged_df.receiver == cleaned_all_companies_df.ibans, "inner") \
    .select(cleaned_all_companies_df.name, final_merged_df.amount, final_merged_df.currency, final_merged_df.timestamp)

# Rename the amount columns to indicate whether it's from sender or receiver
sender_join = sender_join.withColumnRenamed("amount", "sent_amount")
receiver_join = receiver_join.withColumnRenamed("amount", "received_amount")

# Combine sender and receiver data
combined_df = sender_join.join(receiver_join, ["name", "currency", "timestamp"], "outer")

# Handle null values by replacing them with 0
combined_df = combined_df.fillna(0)

# Calculate the total amount for each company
master_table_df = combined_df.groupBy("name", "currency", "timestamp") \
    .agg((F.sum("sent_amount") - F.sum("received_amount")).alias("current_amount"))

# Show the resulting master table
master_table_df.show(truncate=False)
```

```python
from pyspark.sql.functions import col, collect_set, lit

# Step 1: Create a DataFrame with currency-to-country mapping (this is just an example)
currency_country_mapping = [
    ("USD", "United States"),
    ("EUR", "Eurozone"),
    ("INR", "India"),
    ("GBP", "United Kingdom"),
    ("JPY", "Japan")
]

currency_df = spark.createDataFrame(currency_country_mapping, ["currency", "country"])

# Step 2: Join the original DataFrame (final_merged_df) with the currency-country mapping DataFrame
df_with_country = final_merged_df.join(currency_df, final_merged_df.currency == currency_df.currency, "left")

# Step 3: Aggregate and collect distinct countries for each client (name)
countries_by_client = df_with_country.groupBy("name", "currency", "timestamp") \
    .agg(collect_set("country").alias("countries_transacted_with"))

# Step 4: Show the resulting DataFrame
countries_by_client.show(truncate=False)
```
